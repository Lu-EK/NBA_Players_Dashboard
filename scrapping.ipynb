```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd

class get_data:

    def dataset(self, season, mode):
        url = f'https://www.basketball-reference.com/leagues/NBA_{season}_{mode}.html'
        table_web = BeautifulSoup(urlopen(url), 'html.parser').findAll('table')

        df = pd.read_html(str(table_web))[0]  # Changed table_html to table_web
        #df = df.drop(df[df.Player == 'Player'].index)
        #df = df.drop('Rk', axis=1) # drop Rk columns
        #df.Player = df.Player.str.replace('*','',regex=True)
        df.insert(0,'Season',season) # insert season column
        df = df.apply(pd.to_numeric, errors='coerce').fillna(df)

        return df  # Added return statement to return the DataFrame

# Now, to call the dataset function:
dataset_obj = get_data()  # Instantiate the get_dataset object
season = '2022'  # Example season
mode = 'shooting'  # Example mode
result_df = dataset_obj.dataset(season, mode)  # Call the dataset method
result_df.columns = result_df.columns.get_level_values(1)
#result_df.columns = result_df.columns.get_level_values(1)
#result_df = result_df.drop(result_df.index[0])
result_df.reset_index(drop=True, inplace=True)
result_df.drop_duplicates(subset=result_df.columns[2], keep='first', inplace=True)
result_df = result_df.loc[:, ~result_df.columns.str.startswith('Unnamed')]
result_df.reset_index(drop=True, inplace=True)
result_df  # Display the resulting DataFrame

```
